extraction and initial auditing of raw data


```python
import pandas as pd
import numpy as np
import json
```

```python
df = pd.read_csv('../data/raw/nyc_parking_tickets_sample.csv', dtype=str)
df.columns = df.columns.str.replace(' ', '_')
print(f"Shape: {df.shape}")
```

```python
null_report = pd.DataFrame({
    'null_count': df.isnull().sum(),
    'null_pct': (df.isnull().sum() / len(df) * 100).round(2)
})
print(null_report[null_report.null_count > 0])
```

```python
for col in df.columns:
    print(f"{col}: {df[col].nunique()} unique | sample: {df[col].dropna().unique()[:5]}")
```

```python
audit_data = {
    "shape": df.shape,
    "nulls": null_report.to_dict(),
    "dtypes": df.dtypes.astype(str).to_dict()
}
with open('../docs/cleaning_audit.json', 'w') as f:
    json.dump(audit_data, f)
```
